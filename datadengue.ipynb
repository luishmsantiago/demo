```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
arquivo = 'denguecases1620.csv'

# Carregar o arquivo CSV em um DataFrame do Pandas
df = pd.read_csv(arquivo)

df['Year'] = df['Year'].astype(int)

df
```

```python
df = df.groupby(['Year', 'Month']).sum()
df
#Dados somados de casos de dengue por mÃªs e ano em todas as regiÃµes
```

```python
df = df.groupby(['Year', 'Region']).sum()
df
#Dados somados de casos de dengue por regiÃ£o e ano no total
```

```python
import pandas as pd
import matplotlib.pyplot as plt

anos = df.index.tolist()  # Pega os anos (Ã­ndice) como lista
casos = df['Dengue_Cases'].tolist()  # Transforma a coluna em lista
mortes = df['Dengue_Deaths'].tolist()  # Outra coluna em lista

# Exemplo de DataFrame corrigido
data = {
    'Year': anos,
    'Dengue_Cases': casos,
    'Dengue_Deaths': mortes
}

df = pd.DataFrame(data).set_index('Year')  # Define Year como Ã­ndice

# Criar grÃ¡fico
fig, ax = plt.subplots(figsize=(8, 5))

# Criar barras horizontais para Casos e Mortes
df[['Dengue_Cases', 'Dengue_Deaths']].plot(kind='barh', ax=ax, color=['blue', 'red'])

# Configurar labels
ax.set_xlabel('NÃºmero de Casos / Mortes')
ax.set_ylabel('Ano')
ax.set_title('Casos e Mortes por Dengue (2016-2020)')
ax.legend(['Casos de Dengue', 'Mortes por Dengue'])

# Exibir grÃ¡fico
plt.show()
```

```python
plt.pie(casos, labels=anos, autopct='%1.1f%%', startangle=90)
plt.title('DistribuiÃ§Ã£o casos de dengue por ano')
plt.show()
```

```python
plt.pie(mortes, labels=anos, autopct='%1.1f%%', startangle=90)
plt.title('DistribuiÃ§Ã£o mortes de dengue por ano')
plt.show()
```

```python
df = pd.read_csv(arquivo)

df['Year'] = df['Year'].astype(int)
df

df_grouped = df.groupby('Region')[['Dengue_Cases', 'Dengue_Deaths']].sum()
df_grouped = df_grouped.reset_index()  # Traz 'Region' de volta como coluna

print(df_grouped)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Garantir que 'Region' seja o Ã­ndice correto
df_corrigido = df_grouped.set_index('Region')[['Dengue_Cases', 'Dengue_Deaths']]

# Criar grÃ¡fico
fig, ax = plt.subplots(figsize=(8, 5))

# Criar barras horizontais para Casos e Mortes
df_corrigido.plot(kind='barh', ax=ax, color=['blue', 'red'])

# Configurar labels
ax.set_xlabel('NÃºmero de Casos / Mortes')
ax.set_ylabel('RegiÃ£o')  # Agora mostra as regiÃµes corretamente
ax.set_title('Casos e Mortes por Dengue por RegiÃ£o')
ax.legend(['Casos de Dengue', 'Mortes por Dengue'])

# Exibir grÃ¡fico
plt.show()
```
